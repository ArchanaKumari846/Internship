```python
Airline_Name= []
Journey_Date= []
Source= []
Destination= []
Departure_Time= []
Arrival_Time= []
Duration= []
Total_Stops= []
Price= []
links=[]
```

```python
#fetching all details from fetched urls
for l in url:
    driver1.get(l)
    time.sleep(4)
                
    #fetching all Airline_Name
    try:
        Airline=driver1.find_elements(By.XPATH,"//span[@class='i-b text ellipsis']")
        for i in Airline:
            Airline_Name.append(i.text)
    except NoSuchElementException:
        Airline_Name.append('-')
        
      #fetching all Source
    try:
        Source_tag=driver1.find_elements(By.XPATH,"//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p")
        for i in Source_tag:
            Source.append(i.text)
    except NoSuchElementException:
        Source.append('-')
        
    #fetching all Destination
    try:
        Destination_tag=driver1.find_elements(By.XPATH,"//div[@class='i-b pdd-0 text-left atime col-5']/p[2]")
        for i in Destination_tag:
            Destination.append(i.text)
    except NoSuchElementException:
        Destination.append('-')
        
    #fetching all Departure_Time
    try:
        Departure=driver1.find_elements(By.XPATH,"//div[@class='i-b pr']")
        for i in Departure:
            Departure_Time.append(i.text)
    except NoSuchElementException:
        Departure_Time.append('-')

    #fetching all Arrival_Time
    try:
        Arrival_tag=driver1.find_elements(By.XPATH,"//p[@class='bold fs-15 mb-2 pr time']")
        for i in Arrival_tag:
            Arrival_Time.append(i.text)
    except NoSuchElementException:
        Arrival_Time.append('-')

    #fetching all Duration
    try:
        Duration_tag=driver1.find_elements(By.XPATH,"//div[@class='stop-cont pl-13']/p")
        for i in Duration_tag:
            Duration.append(i.text)
    except NoSuchElementException:
        Duration.append('-')

    #fetching all Total_Stops
    try:
        Stops_tag=driver1.find_elements(By.XPATH,"//div[@class='stop-cont pl-13']/div/span")
        for i in Stops_tag:
            Total_Stops.append(i.text)
    except NoSuchElementException:
        Total_Stops.append('-')

    #fetching all Price
    try:
        Price_tag=driver1.find_elements(By.XPATH,"//div[@class='i-b tipsy fare-summary-tooltip fs-18']")
        for i in Price_tag:
            Price.append(i.text)
    except NoSuchElementException:
        Price.append('-')
```

```python
print(len(Airline_Name),len(Source),len(Destination),len(Departure_Time),len(Arrival_Time),len(Duration),len(Total_Stops),len(Price))
```
