# Q.1.Answer: A python program to display all the header tags from wikipedia.org

```python
page= requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

# Q.2.Answer:  A python program to display IMDB’s Top rated 100 movies’ data 

```python
page1= requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&count=100&start=1&ref_=adv_nxt')
page1
```

```python
soup1= BeautifulSoup(page1.content)
soup1
```

```python
name=[]
for i in soup1.find_all('h3', class_="lister-item-header"):
    cells = i.find_all('a')
    name.append(cells[0].text.strip())
name 
```

```python
rating=[]
for i in soup1.find_all('div', class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.replace('\n',''))
rating
```

```python
release=[]
for i in soup1.find_all('span', class_="lister-item-year text-muted unbold"):
    release.append(i.text)
release
```

```python
print('*'*73)
print('\t\t\t IMDB’s Top rated 100 Movies')
print('*'*73)
df_movie=pd.DataFrame({'Name of Movie':name, 'Rating of Movie':rating, 'Year of Release of Movie':release})
df_movie
```

# Q.4.Answer:  A python program to display list of respected former presidents of India

```python
page3= requests.get('https://presidentofindia.nic.in/former-presidents.htm')
print(page3,'\n')

soup3= BeautifulSoup(page3.content)
soup3
```

```python
President_name=[]
for i in soup3.find_all('div', class_="presidentListing"):
    cells = i.find_all('h3')
    President_name.append(cells[0].text.strip())
President_name
```

```python
Term=[]
for i in soup3.find_all('div', class_="presidentListing"):
    cells = i.find_all('p')
    Term.append(cells[0].text.strip().replace('Term of Office:',''))
Term
```

```python
df_president=pd.DataFrame({'Name of President':President_name,'Term of Office':Term})
df_president
```

# Q.5.Answer:  A python program to scrape cricket rankings

```python
page4= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
print(page4,'\n')

soup4= BeautifulSoup(page4.content)
print(soup4,'\n')
```

```python
Team=[]
Matches=[]
Points=[]
Rating=[]
Ranking=[]
for i in soup4.find_all('tr')[:11]:
    cells = i.find_all('td')
    if len(cells) == 5:
        Ranking.append(cells[0].text.strip())
        Team.append(cells[1].text.strip().replace('\n'," , "))
        Matches.append(cells[2].text.strip())
        Points.append(cells[3].text.strip())
        Rating.append(cells[4].text.strip())
df_team = pd.DataFrame({' Ranking':Ranking,'Team':Team,'Matches':Matches,'Points':Points,'Rating':Rating})  
df_team
```

# Q.7.Answer:  A python program to scrape mentioned news details

```python
page7= requests.get('https://www.cnbc.com/world/?region=world')
print(page7,'\n')

soup7= BeautifulSoup(page7.content)

headline=[]
for i in soup7.find_all('a', class_="LatestNews-headline"):
    headline.append(i.get('title'))
print(headline,'\n')

time=[]
for i in soup7.find_all('span', class_="LatestNews-wrapper"):
    time.append(i.text)
print(time,'\n')

News_Link = []
for i in soup7.findAll('a',class_="LatestNews-headline"):
    News_Link.append(i.get('href'))
print(News_Link,'\n')
```

```python
News=pd.DataFrame({'Headline': headline, 'Time': time,'News Link': News_Link})
News
```

# Q.8.Answer: A python program to scrape the details of most downloaded articles from AI in last 90 days. 

```python
page8= requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup8= BeautifulSoup(page8.content)

Paper_Title=[]
for i in soup8.find_all('h2', class_="sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR"):
    Paper_Title.append(i.text)
print(Paper_Title,'\n')

Authors=[]
for i in soup8.find_all('span', class_="sc-1w3fpd7-0 pgLAT"):
    Authors.append(i.text)
print(Authors,'\n')    

Published_Date=[]
for i in soup8.find_all('span', class_="sc-1thf9ly-2 bKddwo"):
    Published_Date.append(i.text)
print(Published_Date,'\n')   

Paper_URL=[]
for i in soup8.find_all('a', class_="sc-5smygv-0 nrDZj"):
    Paper_URL.append(i.get('href'))
print(Paper_URL,'\n')   
```

```python
Articles=pd.DataFrame({'Paper Title': Paper_Title,'Authors': Authors,'Published Date': Published_Date, 'Paper URL':Paper_URL})
Articles
```

# Q.9.Answer:  A python program to scrape  details from dineout.co.in

```python
page9= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup9= BeautifulSoup(page9.content)

restaurant_name= []
for i in soup9.find_all('a', class_="restnt-name ellipsis"):
    restaurant_name.append(i.text)
print(restaurant_name,'\n')

Cuisine=[]
for i in soup9.find_all('div', class_="detail-info"):
    Cuisine.append((i.text.split()[6])+(' ')+(i.text.split()[7]))
print(Cuisine,'\n')


location=[]
for i in soup9.find_all('div', class_="restnt-loc ellipsis"):
    location.append(i.text)
print (location,'\n')

Ratings= []
for i in soup9.find_all('div', class_="restnt-rating rating-4"):
    Ratings.append(i.text)
print(Ratings,'\n')

Image_URL= []
for i in soup9.find_all('img', class_="no-img"):
    Image_URL.append(i['data-src'])
print(Image_URL,'\n')

dineout=pd.DataFrame({'Restaurant_Name':restaurant_name,'Cuisine':Cuisine,'Location':location,'Ratings':Ratings,'Image_URL':Image_URL})
dineout
```

# Q.10.Answer:  A python program to scrape the details of top publications from Google Scholar

```python
page10= requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
soup10= BeautifulSoup(page10.content)

Rank=[]
for i in soup10.find_all('td', class_="gsc_mvt_p"):
    Rank.append(i.text)
print (Rank,'\n')

publication=[]
for i in soup10.find_all('td', class_="gsc_mvt_t"):
    publication.append(i.text)
print (publication,'\n')

h5_index=[]
for i in soup10.find_all('a', class_="gs_ibl gsc_mp_anchor"):
    h5_index.append(i.text)
print (h5_index,'\n')

h5_media=[] 
for i in soup10.find_all('span', class_="gs_ibl gsc_mp_anchor"):
    h5_media.append(i.text)
print (h5_media,'\n')

publications=pd.DataFrame({'Rank':Rank,'Publication':publication,'h5_index':h5_index, 'h5_media':h5_media})
publications
```
